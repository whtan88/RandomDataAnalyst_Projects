# Gemini 日本語の先生

This is a notebook attempting to learn and study for JLPT N5 from Gemini, feeding it the word prompt to act as a Japanese Language Sensei.

Below code is to initialise Gemini generative AI and to provide its configuration settings.

````python
"""
Install the Google AI Python SDK

$ pip install google-generativeai

See the getting started guide for more information:
https://ai.google.dev/gemini-api/docs/get-started/python
"""

!pip install gTTS

from google.colab import userdata
import os

import google.generativeai as genai

# Import additional libraries
from gtts import gTTS
from IPython.display import Audio, display

os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# Create the model
# See https://ai.google.dev/api/python/google/generativeai/GenerativeModel
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  #"max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}
safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
]

model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",
  safety_settings=safety_settings,
  generation_config=generation_config,
  system_instruction="You are a japanese sensei who is teaching a student to prepare for JLPT N5 examination." # japanese teacher and sensei gives different text output formats, with english and without
)
chat_session = model.start_chat(history=[])
````

# Written Conversation Mode

We can also try to have a written conversation with it, and also make it to correct your sentences in case of any errors and to explain it in english.

````python
# Generate a continuous chat conversation until session is killed
while True:
  human_input = input()
  if human_input == "じゃあまた":
    response = chat_session.send_message(human_input)
    print(response.text)
    break
  else:
    response = chat_session.send_message("Have a back and forth conversation in japanese while checking the word prompt for any errors,"
                                         f"and provide correction and english explanation if necessary. {human_input}")
    print(response.text)
````
